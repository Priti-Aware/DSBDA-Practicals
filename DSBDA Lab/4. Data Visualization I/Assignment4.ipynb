```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df
```

```python
# Set the style of seaborn
# sns.set_style("whitegrid")
```

```python
# Plotting survival count based on sex
sns.countplot(x='survived', hue='sex', data=df)
plt.title('Survival Count by Sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# # Survival Count by Sex:
# We see that the number of males who didn't survive (around 470) is much higher than those who survived (around 110).
# In contrast, for females, the number of survivors (around 230) exceeds the number of non-survivors (around 80). 
# This stark difference in survival rates between genders suggests that women were given priority in rescue efforts,
# resulting in a higher survival rate among females.
```

```python
# Plotting survival count based on passenger class
sns.countplot(x='survived', hue='class', data=df)
plt.title('Survival Count by Passenger Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Survival Count by Passenger Class:
# The plot indicates that passengers in the 1st class had a relatively higher chance of survival, with more survivors than non-survivors.
# In contrast, passengers in the 3rd class experienced the opposite scenario, with a significantly higher number of non-survivors compared to survivors.
# This pattern suggests that passenger class played a significant role in determining survival, with higher-class passengers having better 
# access to lifeboats and rescue resources.
```

```python
sns.histplot(df['fare'],bins=20, kde=True)
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# The histogram of ticket prices shows that the distribution is right-skewed, indicating that the majority of ticket prices were relatively low.
# There are few tickets with very high prices, suggesting that these may have been premium or first-class tickets.
# The peak of the distribution lies in the lower fare range, indicating that a large proportion of passengers purchased tickets at lower price points, 
# possibly corresponding to lower-class accommodations.
# The kernel density estimate (KDE) overlay provides a smooth representation of the distribution, highlighting areas of higher and lower density of 
# ticket prices.

# The histogram shows that most tickets were priced on the lower side, with only a few being very expensive. 
# This suggests that many passengers paid less for their tickets, while some opted for higher-priced tickets, 
# likely for better accommodations. Overall, the histogram helps us understand how ticket prices were distributed among Titanic passengers, 
# reflecting differences in affordability and possibly indicating different passenger classes.
```
